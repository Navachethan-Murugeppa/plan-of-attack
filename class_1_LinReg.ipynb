# Importing All packages and Libraries

```python
from sklearn import datasets #contains all the datasets to work within sklearn package

import pandas as pd #scientific libraries to perform operation on dataset
import numpy as np #scientific libraries to perform operation on dataset

from sklearn.linear_model import LinearRegression# Importing algorithms to work for prediction


from sklearn.model_selection import train_test_split # library to divide the data into training set
# and testing set

from sklearn.metrics import mean_squared_error,r2_score #libraries to measure the accuracy and error in prediction

import matplotlib.pyplot as plt # libraries to plot graph and visualize the data distribution 

from sklearn.utils import shuffle #librries to shuffle the data to provide randomization
```

# loading the particular dataset from given set of datasets

```python
boston=datasets.load_boston() #press tab after datasets. to see number of datsets present
```

```python
boston.keys() #to get key items that has certain values(assume like title and description)
```

```python
print(boston['DESCR']) # way to determine the value of particular key
```

```python
X=pd.DataFrame(boston.data,columns=boston.feature_names) # Creating a dataframe to perform operations on data 
X
```

```python
X.describe() #getting overall info of data distribution
```

```python
Y=boston.target # assigning the target value or end feature to be predicted
```

```python
Y
```

# creating new dataframe for single regression i.e how one feature affects target value Y

```python
X_new=pd.DataFrame(X['LSTAT'])
X_new
```

# Simple Regression

creating a model of our algorithm,
dividing the data into training and testing set

```python
model=LinearRegression()
X_train,X_test,Y_train,Y_test = train_test_split(X_new,Y,test_size=0.3,random_state=7)#dividing data into 70% train and 30% test on random seed of 7
X_train
```

```python
model.fit(X_train,Y_train)# give data to model for learning
```

```python
Y_pred=model.predict(X_test)# now asking the model to answer from the learned experience
Y_pred
```

```python
Y_test #actual answer
```

```python
model.score(X_train,Y_train) #determinig how good we predicted on training data
```

```python
model.score(X_test,Y_test) #determining how good we predicted on testing data
```

```python
plt.scatter(X_test['LSTAT'], Y_test,  color='gray') #plotting the points in form of dots through scatter function
plt.plot(X_test['LSTAT'], Y_pred, color='red', linewidth=2)# determinig the best fit line
plt.show()# function to display after pllot creaion on screen
```
